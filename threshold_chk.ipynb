```python
import pandas as pd
from matplotlib.widgets import CheckButtons,Slider
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
# รับข้อมูล
cols_to_use = ['Character', 'Time']
key = pd.read_csv("E:/PyAverage/2 keys/a15.csv",usecols = cols_to_use)
```

```python
# ตัด + เรียงข้อมูล 
key = key[key.Time >= 0]
key = key[key.Time <= 10000]
key = key.sort_values(by=['Character'])
```

```python
# จัดกลุ่มข้อมูล Character และหาค่าเฉลี่ย
means = key.groupby('Character')['Time'].mean()
means = pd.DataFrame(means)
# means
```

```python
# หา % ค่าเฉลี่ย
Per5 = means['Time'] * 0.05
Per10 = means['Time'] * 0.1
Per15 = means['Time'] * 0.15
Per20 = means['Time'] * 0.2
```

```python
# สร้างตารางเก็บ % และ ค่าเฉลี่ย
ts = pd.DataFrame()
ts['average'] = means['Time']
ts['5%'] = Per5
ts['+5%'] = ts['average'] + ts['5%']
ts['-5%'] = ts['average'] - ts['5%']

ts['10%'] = Per10
ts['+10%'] = ts['average'] + ts['10%']
ts['-10%'] = ts['average'] - ts['10%']

ts['15%'] = Per15
ts['+15%'] = ts['average'] + ts['15%']
ts['-15%'] = ts['average'] - ts['15%']

ts['20%'] = Per20
ts['+20%'] = ts['average'] + ts['20%']
ts['-20%'] = ts['average'] - ts['20%']
# ts
```

```python
ch = pd.DataFrame()

ch = ts[(ts.average >= 10000) | (ts.average < 0) |
       (ts['5%'] >= 10000) | (ts['5%'] < 0) |
       (ts['+5%'] >= 10000) | (ts['+5%'] < 0) |
       (ts['-5%'] >= 10000) | (ts['-5%'] < 0) |
       (ts['10%'] >= 10000) | (ts['10%'] < 0) |
       (ts['+10%'] >= 10000) | (ts['+10%'] < 0) |
       (ts['-10%'] >= 10000) | (ts['-10%'] < 0) |
       (ts['15%'] >= 10000) | (ts['15%'] < 0) |
       (ts['+15%'] >= 10000) | (ts['+15%'] < 0) |
       (ts['-15%'] >= 10000) | (ts['-15%'] < 0) |
       (ts['20%']>= 10000) | (ts['20%'] < 0) |
       (ts['+20%'] >= 10000) | (ts['+20%'] < 0) |
       (ts['-20%'] >= 10000) | (ts['-20%'] < 0)]
ch
```
