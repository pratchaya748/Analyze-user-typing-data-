```python
import pandas as pd
from matplotlib.widgets import CheckButtons
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure

import ipywidgets as wg
from IPython.display import display
%matplotlib notebook
from ipywidgets import Label, Layout, HBox
```

```python
# ไฟล์นำมาคำนวนค่าเฉลี่ย,thrsehold
cols_to_use = ['Character', 'Time']
key = pd.read_csv("2 keys/a1.csv", usecols = cols_to_use)

key
```

```python
# รับข้อมูล test
test = pd.read_csv("2 keys/a10.csv",usecols = cols_to_use)

test
```

```python
# ตัด + เรียงข้อมูล 
key = key[key.Time >= 0]
key = key[key.Time <= 10000]
key = key.sort_values(by=['Character'])

# จัดกลุ่มข้อมูล Character และหาค่าเฉลี่ย
means = key.groupby('Character')['Time'].mean()
means = pd.DataFrame(means)
means
```

```python
# ตัด + เรียงข้อมูล test
test = test[test.Time >= 0]
test = test[test.Time <= 10000]
test = test.sort_values(by=['Character'])

# สุ่มตัวอย่างของ test
test = test.groupby("Character").sample(n=1, random_state=1)
test
```

```python
# หา % ค่าเฉลี่ย
Per5 = means['Time'] * 0.05
Per10 = means['Time'] * 0.1
Per15 = means['Time'] * 0.15
Per20 = means['Time'] * 0.2

# สร้างตารางเก็บ % และ ค่าเฉลี่ย
ts = pd.DataFrame()
ts['average'] = means['Time']
ts['5%'] = Per5
ts['+5%'] = ts['average'] + ts['5%']
ts['-5%'] = ts['average'] - ts['5%']

ts['10%'] = Per10
ts['+10%'] = ts['average'] + ts['10%']
ts['-10%'] = ts['average'] - ts['10%']

ts['15%'] = Per15
ts['+15%'] = ts['average'] + ts['15%']
ts['-15%'] = ts['average'] - ts['15%']

ts['20%'] = Per20
ts['+20%'] = ts['average'] + ts['20%']
ts['-20%'] = ts['average'] - ts['20%']
ts = ts.reset_index()
```

```python
# ตัด Character
test = test.loc[test['Character'].isin(ts['Character'])]
test
```

```python
check_user = pd.DataFrame()

# เก็บค่าที่อยู่นอกหรือใน teshold
checker = []
check_user['Character'] = test['Character'] 
    
# แปลงเป็น dic
data = dict(zip(test['Character'],test['Time']))

t5 = dict(zip(ts.Character,ts['+5%']))
b5 = dict(zip(ts.Character,ts['-5%']))

t10 = dict(zip(ts.Character,ts['+10%']))
b10 = dict(zip(ts.Character,ts['-10%']))

t15 = dict(zip(ts.Character,ts['+15%']))
b15 = dict(zip(ts.Character,ts['-15%']))

t20 = dict(zip(ts.Character,ts['+20%']))
b20 = dict(zip(ts.Character,ts['-20%']))

# ฟังก์ชันวนเช็คค่า
def check5():
    for k,v in data.items():
        if v > t5[k[:2]] or v < b5[k[:2]]:
            checker.append(0)
        else :
            checker.append(1)
            
def check10():
    for k,v in data.items():
        if v > t10[k[:2]] or v < b10[k[:2]]:
            checker.append(0)
        else :
            checker.append(1)

def check15():
    for k,v in data.items():
        if v > t15[k[:2]] or v < b15[k[:2]]:
            checker.append(0)
        else :
            checker.append(1)
            
def check20():
    for k,v in data.items():
        if v > t20[k[:2]] or v < b20[k[:2]]:
            checker.append(0)
        else :
            checker.append(1)

# เรียกฟังก์ชัน
check20()
check_user['checker'] = checker

check_user
```

```python
# นับจำนวนที่อยู่ในและนอกteshold
# 1 = in >> real user
# 0 = out >> fake user

num = check_user['checker'].to_list()
num1 = 0
num0 = 0
for i in num:
    if i== 1:
        num1 += 1
    elif i == 0:
        num0 += 1

user = num1 / (num1+num0)*100

if user > 50:
    result = "real user"
elif user == 50:
    result = "unidentify"
elif user < 50:
    result = "fake user"


print("true = ",num1)
print("false = ",num0)    
print(f'{user:.2f}% = {result}')
```

```python
# dataframe สำหรับทำกราฟ
dfThreshold = pd.DataFrame()

# เก็บค่า
dfThreshold['Character'] = test['Character']
lAvg = []
lt5 = []
lb5 = []
lt10 = []
lb10 = []
lt15 = []
lb15 = []
lt20 = []
lb20 = []

# แปลงเป็น dic
avg = dict(zip(ts.Character,ts['average']))

t5 = dict(zip(ts.Character,ts['+5%']))
b5 = dict(zip(ts.Character,ts['-5%']))

t10 = dict(zip(ts.Character,ts['+10%']))
b10 = dict(zip(ts.Character,ts['-10%']))

t15 = dict(zip(ts.Character,ts['+15%']))
b15 = dict(zip(ts.Character,ts['-15%']))

t20 = dict(zip(ts.Character,ts['+20%']))
b20 = dict(zip(ts.Character,ts['-20%']))


for k,v in data.items():
    lAvg.append(avg[k[:2]])
    lt5.append(t5[k[:2]])
    lb5.append(b5[k[:2]])
    
    lt10.append(t10[k[:2]])
    lb10.append(b10[k[:2]])
    
    lt15.append(t15[k[:2]])
    lb15.append(b15[k[:2]])
    
    lt20.append(t20[k[:2]])
    lb20.append(b20[k[:2]])

dfThreshold['Avg'] = lAvg
dfThreshold['Per_T_5'] = lt5
dfThreshold['Per_B_5'] = lb5

dfThreshold['Per_T_10'] = lt10
dfThreshold['Per_B_10'] = lb10

dfThreshold['Per_T_15'] = lt15
dfThreshold['Per_B_15'] = lb15

dfThreshold['Per_T_20'] = lt20
dfThreshold['Per_B_20'] = lb20

dfThreshold
```

```python
# ฟังก์ชันไว้คำนวน %
def pec_check(functionCheck):
    checker2 = []
    
    # ฟังก์ชันวนเช็คค่า
    def check5():
        for k,v in data.items():
            if v > t5[k[:2]] or v < b5[k[:2]]:
                checker2.append(0)
            else :
                checker2.append(1)

    def check10():
        for k,v in data.items():
            if v > t10[k[:2]] or v < b10[k[:2]]:
                checker2.append(0)
            else :
                checker2.append(1)

    def check15():
        for k,v in data.items():
            if v > t15[k[:2]] or v < b15[k[:2]]:
                checker2.append(0)
            else :
                checker2.append(1)

    def check20():
        for k,v in data.items():
            if v > t20[k[:2]] or v < b20[k[:2]]:
                checker2.append(0)
            else :
                checker2.append(1)

    # เรียกฟังก์ชัน
    eval(functionCheck)

    
    num = checker2
    num1 = 0
    num0 = 0
    for i in num:
        if i== 1:
            num1 += 1
        elif i == 0:
            num0 += 1

    user = num1 / (num1+num0)*100

    if user > 50:
        result = "real user"
    elif user == 50:
        result = "unidentify"
    elif user < 50:
        result = "fake user"


    print("true = ",num1)
    print("false = ",num0)    
    print(f'{user:.2f}% = {result}')
    
# pec_check("check15()")
# pec_check("check10()")
```

```python
output1 = wg.Output()
output2 = wg.Output()
output3 = wg.Output()
output4 = wg.Output()
display(HBox([output1,output2,output3,output4]))

def threshold(Per5 = False ,Per10 = False ,Per15 = False ,Per20 = True):
    fig, ax = plt.subplots(figsize=(10,5))
    plt.scatter(dfThreshold['Character'],dfThreshold['Avg'])
    plt.scatter(test['Character'],test['Time'],marker='+')
    
    plt.title("2keys Average Time")
    plt.xlabel("Character")
    plt.ylabel("Time : ms")
    
    output2.clear_output()
    output1.clear_output()
    output3.clear_output()
    output4.clear_output()
    
    if Per5 == True:
        plt.plot(dfThreshold['Character'],dfThreshold['Per_T_5'],color="#E00296")
        plt.plot(dfThreshold['Character'],dfThreshold['Per_B_5'],color="#E00296")
        with output1:
            print('threshold 5%')
            pec_check("check5()")
            print("--------------------")
        
    if Per10 == True:
        plt.plot(dfThreshold['Character'],dfThreshold['Per_T_10'],color="red")
        plt.plot(dfThreshold['Character'],dfThreshold['Per_B_10'],color="red")
        with output2:
            print('threshold 10%')
            pec_check("check10()")
            print("--------------------")
        
        
    if Per15 == True:
        plt.plot(dfThreshold['Character'],dfThreshold['Per_T_15'],color="orange")
        plt.plot(dfThreshold['Character'],dfThreshold['Per_B_15'],color="orange")
        with output3:
            print('threshold 15%')
            pec_check("check15()")
            print("--------------------")
        
        
    if Per20 == True:
        plt.plot(dfThreshold['Character'],dfThreshold['Per_T_20'],color="green")
        plt.plot(dfThreshold['Character'],dfThreshold['Per_B_20'],color="green")
        with output4:
            print('threshold 20%')
            pec_check("check20()")
            print("--------------------")
        
        
    plt.show()

wg.interact(threshold ,Per5 = False ,Per10 = False ,Per15 = False ,Per20 = True)
```
